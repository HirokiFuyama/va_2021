```python
def read_audio(file_path=a_path):
    """
    :return:List of array of raw data.
    """
    file_path.sort()
    raw_data = [sf.read(file_path[i])[0] for i in range(len(file_path))]
    return raw_data
```

```python
def check_length(data):
    fs = 44100
    return len(data)/fs
```
